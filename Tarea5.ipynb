### 5.5.2. Calculando O

1.   $5n + 7$<BR>
Complejidad $O(n)$

```python
import matplotlib.pyplot as plt
import numpy as np

n = np.linspace(0, 10, 100)

f_n = 5 * n + 7

plt.figure(figsize=(8, 6))
plt.plot(n, f_n, label='5n + 7', color='blue')
plt.xlabel('n')
plt.ylabel('f(n)')
plt.title('5n + 7')
plt.legend()
plt.grid(True)
plt.show()
```

2. $ 7n^2 + 3n - 7 $<BR>
Complejidad $O(n^2)$[texto del enlace](https://)

```python
import matplotlib.pyplot as plt
import numpy as np

n = np.linspace(0, 10, 100)

f_n = 7*n*n+3*n-7

plt.figure(figsize=(8, 6))
plt.plot(n, f_n, label='$ 7n^2 + 3n - 7 $', color='blue')
plt.xlabel('n')
plt.ylabel('f(n)')
plt.title('$ 7n^2 + 3n - 7 $')
plt.legend()
plt.grid(True)
plt.show()
```


3. $2n^3 + 7^{23} n + 2$<BR>
Complejidad $O(n^3)$

```python
import matplotlib.pyplot as plt
import numpy as np

n = np.linspace(0, 10)

f_n = 2* n**3 + 7**23 * n + 2

plt.figure(figsize=(8, 6))
plt.plot(n, f_n, label='$2n^3 + 7^{23} n + 2$', color='blue')
plt.xlabel('n')
plt.ylabel('f(n)')
plt.title('$2n^3 + 7^{23} n + 2$')
plt.legend()
plt.grid(True)
plt.show()
```
