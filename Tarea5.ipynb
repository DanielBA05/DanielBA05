### 5.5.2. Calculando O

1.   $5n + 7$<BR>
Complejidad $O(n)$

```python
import matplotlib.pyplot as plt
import numpy as np

n = np.linspace(0, 10, 100)

f_n = 5 * n + 7

plt.figure(figsize=(8, 6))
plt.plot(n, f_n, label='5n + 7', color='blue')
plt.xlabel('n')
plt.ylabel('f(n)')
plt.title('5n + 7')
plt.legend()
plt.grid(True)
plt.show()
```

2. $ 7n^2 + 3n - 7 $<BR>

```python
import matplotlib.pyplot as plt
import numpy as np

n = np.linspace(0, 10, 100)

f_n = 7*n*n+3*n-7

plt.figure(figsize=(8, 6))
plt.plot(n, f_n, label='$ 7n^2 + 3n - 7 $', color='blue')
plt.xlabel('n')
plt.ylabel('f(n)')
plt.title('$ 7n^2 + 3n - 7 $')
plt.legend()
plt.grid(True)
plt.show()
```


3. $2n^3 + 7^{23} n + 2$<BR>

4. $\frac{n}{2000!} +9 $<BR>
Complejidad $O(1)$<BR>
Este es un ejemplo más pequeño porque para 2000 crecía demasiado

```python
import matplotlib.pyplot as plt
import numpy as np
from math import factorial
from decimal import Decimal

def funcion(n):
    return Decimal(n) / Decimal(factorial(20)) + 9

n_values = np.linspace(0, 1000)

f_values = np.array([funcion(n) for n in n_values])

plt.figure(figsize=(8, 6))
plt.plot(n_values, f_values, label=r'$f(n) = \frac{n}{20!} + 9$') # Fixed the LaTeX expression in the label
plt.title(r"$\frac{n}{2000!} +9$") # Fixed the LaTeX expression in the title and used a raw string
plt.xlabel("n")
plt.ylabel("f(n)")
plt.legend()
plt.grid(True)
plt.show()
```

5. $\frac{7n^4}{2n} + 2^9n^2+8n+3$<BR>
Complejidad $O(n^3)$

```python
import matplotlib.pyplot as plt
import numpy as np

n = np.linspace(0, 10, 100)

f_n = ((7*n**4)/(2*n + 1e-8)) + 2**9*n*n+8*n+3

plt.figure(figsize=(8, 6))
plt.plot(n, f_n, label=r'$\frac{7n^4}{2n} + 2^9n^2+8n+3$', color='blue')
plt.xlabel('n')
plt.ylabel('f(n)')
plt.title(r'$\frac{7n^4}{2n} + 2^9n^2+8n+3$')
plt.legend()
plt.grid(True)
plt.show()
```

7. $2\sqrt{n}+8$<BR>
Complejidad $O(\sqrt{n})$

```python
import matplotlib.pyplot as plt
import numpy as np

n = np.linspace(0, 10, 100)

f_n = 2*np.sqrt(n) + 8

plt.figure(figsize=(8, 6))
plt.plot(n, f_n, label=r'$2\sqrt{n}+8$', color='blue')
plt.xlabel('n')
plt.ylabel('f(n)')
plt.title(r'$2\sqrt{n}+8$')
plt.legend()
plt.grid(True)
plt.show()
```

8. $\log_{2}{(n)}^n+100!$<BR>
Complejidad $O(n\log{n})$

```python
import matplotlib.pyplot as plt
import numpy as np
import math
n = np.linspace(0, 10, 100)

f_n = np.abs(np.log2(n + 1e-8))**n + math.factorial(100)

plt.figure(figsize=(8, 6))
plt.plot(n, f_n, label=r'$\log_{2}{(n)}^n+100!$', color='blue')
plt.xlabel('n')
plt.ylabel('f(n)')
plt.title(r'$\log_{2}{(n)}^n+100!$')
plt.legend()
plt.grid(True)
plt.show()
```

9. $2^nn^3+n!$<BR>
Complejidad $O(n!)$

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.special import factorial

n = np.linspace(0, 10, 100)

f_n = 2**n * n**3 + factorial(n)

plt.figure(figsize=(8, 6))
plt.plot(n, f_n, label=r'$2^nn^3+n!$' , color='blue')
plt.xlabel('n')
plt.ylabel('f(n)')
plt.title(r'$2^nn^3+n!$')
plt.legend()
plt.grid(True)
plt.show()
```

10. $\frac{2^n}{n} + \sum_{i=0}^{100} (i + 1) n^i$<BR>
Complejidad $O(\frac{2^n}{n})$

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.special import factorial

n = np.linspace(0, 10, 100)
def sum(n,i,res):
    if i>100:
        return res
    return sum(n,i+1,res+(i+1*n**i))
f_n = 2**n / (n + 1e-8) + sum(n,0,0)

plt.figure(figsize=(8, 6))
plt.plot(n, f_n, label=r'$\frac{2^n}{n} + \sum_{i=0}^{100} (i + 1) n^i$' , color='blue')
plt.xlabel('n')
plt.ylabel('f(n)')
plt.title(r'$\frac{2^n}{n} + \sum_{i=0}^{100} (i + 1) n^i$')
plt.legend()
plt.grid(True)
plt.show()
```
